# Breakdown of Terminology w/ a Chart

```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
def random_plots(data, time_axis, measurements_axis, title):
    plot_ts = pd.DataFrame(data, columns=["Daily Steps Observations"])
    # print(plot_ts.columns)
    ax = plot_ts.plot(style='.-')
    ax.set(title=title, xlabel=time_axis, ylabel=measurements_axis)
    return plot_ts
```

```python
def get_time():
    time_frame = pd.date_range(start='6/1/2022', end='6/30/2022')
    # print(type(time_frame))
    
    days_in_time_frame = time_frame.max() - time_frame.min() 
    # print(type(days_in_time_frame))
    
    convert_days_in_tf = days_in_time_frame.days + 1
    # print(type(convert_days_in_tf), convert_days_in_tf)
    
    return convert_days_in_tf
```

```python
time = get_time()
data = np.random.randint(1000, 2000, time)
x_axis = 'Time (days)'
y_axis = 'Number of Steps'
title = 'Time Series for Steps Phenomenon'
time_series_data = random_plots(data, x_axis, y_axis, title)
```

```python
# print(type(time_series_data))
def create_df(ts_data):
    ts_data = pd.DataFrame(time_series_data)
    return ts_data
```
