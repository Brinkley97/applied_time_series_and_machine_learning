### Load Raw Data

- **NOTE**: TS frequency is 1 hour

```python
# Only grab stocks whose data is available for the entire time period
start_date, end_date = "2022-11-10", "2023-11-10"
Stock = namedtuple("Stock", ["symbol", "name"])
stocks = [
    ("MSFT", "Microsoft"),
    ("INTC", "Intel")
]
independent_variable = "Close"
stocks = [Stock(*s) for s in stocks]
stocks = {s.symbol: build_stock_uts(s.symbol, s.name, independent_variable, start_date=start_date, end_date=end_date, frequency='1h') for s in stocks}
```

```python
values_cols = list(stocks.keys())
stock_mvts = TimeSeriesFactory.create_time_series(
    time_col="date",
    time_values=stocks[values_cols[0]].data.index,
    values_cols=values_cols,
    values=[stock.get_series() for stock in stocks.values()]
)
```

### Initialize Autoregressive Integrated Moving Average ARIMA(p, d, q) Model Without My Library

- **NOTE**: we use the original TS and NOT the differences TS. Why? The differencing is being handled in our ARIMA model.

```python
import matplotlib.pyplot as plt

from math import sqrt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
true_labels = stocks[stock_symbol].get_series()
true_labels
```

```python
model = ARIMA(true_labels, order=(1, 1, 8))
result = model.fit()
```

```python
print(result.summary())
```

### Predict Forecasts for Raw TS Without My Library

```python
end = 60
# end = len(stock_df)
end
```

```python
predictions = result.predict(start=1, end=end, dynamic=False)
predictions
```

```python
true_labels_subset = true_labels[:end]
len(true_labels_subset), true_labels_subset
```

```python
mse = mean_squared_error(true_labels_subset, predictions)
mse
```

```python
rmse = sqrt(mean_squared_error(true_labels_subset, predictions))
rmse
```

### Plot Actual Forecasts vs Predicted Forecasts for Raw TS Without My Library

```python
plt.figure(figsize=(20, 4))
plt.xlabel("Observations")
plt.ylabel("Values")

plt.plot(true_labels_subset, color='blue', label='Actual Forecasts', linewidth=4)
plt.plot(predictions, color='red', label='Predicted Forecasts', linewidth=1)
```

### Initialize Autoregressive Integrated Moving Average ARIMA(p, d, q) Model With My Library

- With my library to confirm results.
- **NOTE**: we use the original TS and NOT the differences TS. Why? The differencing is being handled in our ARIMA model.

```python
arima_model = ARIMA_model()
```

```python
trained_arima_models = arima_model.train_arima_model(true_labels, 1, 1, 8)
trained_arima_models
```

### Predict Forecasts for Raw TS With My Library

```python
my_predictions = arima_model.predict(trained_arima_models, 1, end)
my_predictions
```

```python
EvaluationMetric.eval_mse(true_labels_subset, my_predictions)
```

```python
EvaluationMetric.eval_rmse(true_labels_subset, my_predictions)
```

### Plot Actual Forecasts vs Predicted Forecasts for Raw TS With My Library
