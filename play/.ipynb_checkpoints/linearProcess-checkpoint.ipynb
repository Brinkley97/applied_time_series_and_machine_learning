# 3.1 General Linear Process

$$
\tilde{z}_t = a_t + \psi_1a_{t - 1} + \psi_2a_{t - 2} ... \newline
 = a_t + \sum_{j = 1}^{\infty} \psi_ja_{t - j}
$$

# Imports

```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
```

# Small Sample
- Remove comments to see the process

```python
def sum_weights_by_white_noise(a, psi, t, js):
    """
    t -- int
    """
    sum_weight_by_noise = 0
    
    a_t = a[t]
    # print(a_t)
    
    for j in range(1, js):
        # print("t:", t, "j:", j)
        # print("t - j")
        # print(t, "-", j, "so, a_t @", t - j, "is", a[t - j])
        # print(psi[j], "*", a[t - j])
        sum_weight_by_noise += psi[j] * a[t - j]
        # print(psi[j], "*", a[t - j], "=", sum_weight_by_noise)
        # sum_weight_by_noise += sum_weight_by_noise
    return sum_weight_by_noise
```

```python
def generate_linear_process(a, psi):
    
    ts = len(a)
    js = len(psi)

    z_ts = []

    for t in range(ts):
        # print("t:", t)
        weighted_sum_of_white_noise = sum_weights_by_white_noise(a, psi, t, js)    
        z_t = weighted_sum_of_white_noise + a[t]
        # print("weighted_sum_of_white_noise: ", weighted_sum_of_white_noise, "+", a[t], "=", z_t)
        z_ts.append(z_t)
        # print()
    return z_ts
```

```python
white_noise_a = [8, 2, 3, 1]
weights_psi = [9, 4, 5]

glp = generate_linear_process(white_noise_a, weights_psi)
plt.plot(glp)
plt.show()
```

# Random Small Sample

```python
t = np.random.randint(low=1, high=5)
# print("t: ", t, t+1)

# t + 1 bc if low can't = high as in if low is 1, high can't be 1; high must be at least two
j = np.random.randint(low=1, high=t+1)
# print("j: ", j)

white_noise_a = np.random.random(t)
weights_psi = np.random.random(j)

glp = generate_linear_process(white_noise_a, weights_psi)
plt.plot(glp)
```
