## 3. More Complex Figures

1. Heat dynamics of a building [4a |P 2/19]
2. Spectral density for a synthetic TS [4c]

---

1. Measurements from an unoccupied test building shown every 12 hours
    - Characterize the thermal behavior of the building
    - Upper plot : ambient air temperature, heat supply, and solar radiation (as inputs/explanatory variables)
    - Lower plot : indoor air temperature (as output/dependent variables)
    - Methods for transfer function modeling as described in
        - CH 8 : Linear systems and stochastic processes



2. Spectral density for a synthetic TS
    - Characteristics 
        - Prescribed periods of 20 and 80
        - Polynomial trend of order 2
        - White noise/signal ratio of 0.5
        - 40% missing values
        - Outliers 5 to 7 standard deviation away from the mean
    - Methods in use : 
        - MTM
        - WWZ
        - Lomb-Scargle
