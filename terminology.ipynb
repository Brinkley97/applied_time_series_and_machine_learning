# Terminology

- **Sensor(s)** : the instrument(s) (also known as hardware(s), software(s), device(s)) used to capture the measurements of some TS phenomena
    - **Sensor Readings** : either at a continuous or discrete *time*, it's the process in which the sensor(s) captures and stores measurements
        - **Continuous** : *at any time*, we can obtain an observation
            - sensor readings occur every second
            - of units of time, sensor readings occur at any unit time
            - Ex :
                - Real #s of times 
                - 9.9 secs, mins, days, weeks, months, etc
            - will have more data bc sensor readings occur more often
        - **Discrete** : we can only obtain an observation *at specific times*
            - sensor readings occur greater than every second
            - of units of time, sensor readings occur at specific units of time
            - Ex :
                - Whole #s of time only 
                - 9 seconds, mins, days, weeks, months, etc
                - 9 mins and 9 secs? yes, but not 9.1 mins and 9.8 seconds
                - 25 years 4 months? yes, but not 25 1/2 years and 4 1/2 months
             - will have less data bc sensor readings occur less often
             
- **TS Dataset** : a collection of TS data
    - **TS** : data that's an ordered set of observations moving through continuous or discrete time
    
- **Observation** : *a single point* along both the x-axis (timestamps) and y-axis (measurements), hence each observation contains a time and some numerical value
    - **Observations** : *more than one* **observation**; a list of either entire dataset or sliced between timestamps to get specific measurements

- **Phenomena** : the observed variables

- **Timestamp** : *a single point* in time in which the sensor captured a measurement from the phenomena; x-axis
    - **Timestamps** : *more than one* **timestamp**; a list of either entire dataset or sliced between timestamps

- **Data Quality** : the condition in which the measurements are presented. $ \exists $ errors (ie : misreadings, NaN, negative #s when all should be positive)? 
    - How much data cleaning should we do?
    - **Dataset Quality** : the condition in which the dataset itself is presented. What file formats can we load with? Do we have to convert the file format or reduce the size of the file?
        
- **Measurement** : *a single numerical value* within the observed phenomenon which is a real #; y-axis
    - **Measurements** : *more than one* **measurement**; a list of either entire measurements or sliced between measurements
    - Either continuous or discrete
        - **Continuous** : 
            - There are an **infinite number of possible values between any two values** [5.1g | 24/28];
                - has no end
            - A phenomenon whose value is *obtained by measuring* - can measure at anytime
            - Any numeric value - **real #s**
                - *divided into smaller increments*
            - Ex :
                - temperature, weight, height, age, blood sugar level when checking diabetes, stock value/price per share, grade in course
            - Think : "If I was to count the #, how long would it take?"

        - **Discrete** : 
            - There are a **finite number of possible values between any two values** [5.1g | 29/33];
                - has an end
            - The phenomenon has a *countable number* of possible values - can count the specific #
            - Counts are nonnegative integers (so ONLY **whole #s**) 
                - *cannot be divided into smaller increments*
            - Ex :
                - #people in a room, #cars in parking lot, #calls in call center, #buildings on campus, #retail/sales transactions, #email advertisements from WeBuyBlack, #earthquakes in SolCal, #stock I have, #heart beats when exercising, #students in class, #students in research lab
            - Think : "If I was to count the #, how long would it take?" and "number of..."
     - **Real #** : any possible # which includes
        - **Natural** : whole #s; ie : 1, 882, 4797, N, so only positive #s
        - **Integers** : **natural** (or positive) #s along w/ negative #s; ie : -N, -4797, -882, -1, 0, 1, 882, 4797, N
        - **Rational** : can be divided so fractions/decimal-based #s
  
- **Frequency** : 
    - Refers to the time interval between the observations of a TS [5.1d]
    - ~Rate at which something happens or is repeated [5.1h]~
    - Returns a numerical value with a unit of time
    - Ex [5.1d]
    
- **Periodicity** : 
    - Observations occur at what unit of time (secs, mins, hrs, etc)
    - Returns the unit of time
    - Ex on [5.1a | P 240/258]

- **Sampling** : occurs every so often

- **Percent** : % = (f / n) x 100, where n is a sum all the phenomenons

- **Pattern** : see below
